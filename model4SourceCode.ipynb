```python
# Importing the libraries
import pandas as pd
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Input
from tensorflow_addons.metrics import RSquare
from sklearn.model_selection import train_test_split
```

```python
features_train = pd.read_csv ('./model4FeaturesTrained.csv', index_col=0)
labels_train = pd.read_csv ('./model4LabelsTrained.csv', index_col=0)
features_test = pd.read_csv ('./model4FeaturesTested.csv', index_col=0)
labels_test = pd.read_csv ('./model4LabelsTested.csv', index_col=0)
```

```python
# https://github.com/katanaml/sample-apps/blob/master/04/multi-output-model.ipynb
def build_model():
    # Define model layers.
    input_layer = Input(shape=(len(features_train.columns),))
    first_dense = Dense(units='128', activation='linear')(input_layer)
    # Y1 output will be fed from the first dense
    y1_output = Dense(units='1', name=labels_train.columns[0])(first_dense)
    second_dense = Dense(units='128', activation='linear')(first_dense)
    # Y2 output will be fed from the second dense
    y2_output = Dense(units='1', name=labels_train.columns[1])(second_dense)
    third_dense = Dense(units='128', activation='linear')(second_dense)
    # Y3 output will be fed from the second dense
    y3_output = Dense(units='1', name=labels_train.columns[2])(third_dense)
    # Define the model with the input layer and a list of output layers
    model = Model(inputs=input_layer, outputs=[y1_output, y2_output, y3_output])
    return model
```

```python
model = build_model()
# Specify the optimizer, and compile the model with loss functions for both outputs
optimizer=tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=optimizer,
              loss={
                  labels_train.columns[0]: 'mse', 
                  labels_train.columns[1]: 'mse',
                  labels_train.columns[2]: 'mse'
              },
              metrics={

                  labels_train.columns[0]: [
                      tf.keras.metrics.RootMeanSquaredError(),
                      tf.keras.metrics.mean_absolute_error,
                      RSquare()],
                  labels_train.columns[1]: [
                      tf.keras.metrics.RootMeanSquaredError(),
                      tf.keras.metrics.mean_absolute_error,
                      RSquare()],
                  labels_train.columns[2]: [
                      tf.keras.metrics.RootMeanSquaredError(),
                      tf.keras.metrics.mean_absolute_error,
                      RSquare()
                  ],
              })
# Fit model (same as above)
history = model.fit(features_train, labels_train, epochs=100) 
# commented out on purpose (not fitting it just yet)
```

```python
trainingHistory = pd.DataFrame(history.history)

from pathlib import Path
                                    
filepath = Path('./model4TrainingHistory.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
trainingHistory.to_csv(filepath) 

trainingHistory 
```

```python
output_preds = model.predict(features_test)
```

```python
predictionArrayTemp = np.array(output_preds)

predictionArray = np.transpose(predictionArrayTemp)
predictionArray[0].shape

prediction_df = pd.DataFrame(predictionArray[0], index=features_test.index, columns=labels_test.columns)

prediction_df
```

```python
albertPrediction = pd.DataFrame(index=features_test.index)

albertPrediction['realClose_nextDay'] = labels_test['Close_nextDay']
albertPrediction['Predict_nextDay'] = prediction_df['Close_nextDay']


albertPrediction['realClose_next6mth'] = labels_test['Close_next_6mth']
albertPrediction['Predict_next6mth'] = prediction_df['Close_next_6mth']


albertPrediction['realClose_next1year'] = labels_test['Close_next_1yr']
albertPrediction['Predict_next1year'] = prediction_df['Close_next_1yr']

albertPrediction
```

```python
from pathlib import Path
                                    
filepath = Path('./model4PredictionResult.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
albertPrediction.to_csv(filepath) 
```
